```python
from sympy import randprime
from datetime import datetime

x,y = (3, 6)

a = 2
b = 3

p_order = 1e8
lower_bound = 1
upper_bound = 10

searched = 600

passed = 0

search_higher = 1

while True:

    p = randprime(lower_bound*p_order, upper_bound*p_order)

    if upper_bound - lower_bound < 0.2 or p is None:
        print(f"\n\nБлижайшие к времени поиска p для p_order = {p_order} найдены!\n\n")
        # search_higher = int(input("Введите 0, чтобы остановиться: "))
        if search_higher:
            p_order *= 10
            lower_bound = 1
            upper_bound = 10
            continue
        break

    print(f"Выполняем для p = {p}")

    cur_x, cur_y = x,y

    order = 1

    start = datetime.now()
    
    while True:
        order += 1
        if cur_x == x:
            if (cur_y + y) % p == 0:
                break
            m = (3 * x * x + a) * pow(2 * y, p-2, p)
        else:
            m = (cur_y - y) * pow(cur_x - x, p-2, p)
        
        new_x = (m * m - cur_x - x) % p
        new_y = (-y - m * (new_x - x)) % p
        
        cur_x, cur_y = new_x, new_y

    passed = (datetime.now() - start).total_seconds()
    print(f"Время подбора {passed} секунд, порядок: {order}\n")
    
    if (passed < searched):
        lower_bound = p / p_order

    if (passed > searched):
        upper_bound = p / p_order
```

```python
import time
from ecdsa import ellipticcurve, numbertheory

def find_point_order(curve, point, p):
    """Находит порядок точки полным перебором."""
    start_time = time.time()
    Q = point
    order = 1
    while True:
        if Q.x() is None and Q.y() is None:  # Бесконечно удаленная точка
            break
        order += 1
        Q = Q + point
        if time.time() - start_time > 600:  # Превышение 10 минут
            raise TimeoutError("Превышено время выполнения (10 минут)")
    return order

# Выбираем простое p ~ 10^5
p = 100003  # Простое число
a = 1
b = 1

# Создаем кривую над Z_p
curve = ellipticcurve.CurveFp(p, a, b)

# Выбираем точку на кривой (подбираем вручную)
# Перебираем x, пока не найдем такой, что x^3 + a*x + b — квадрат по модулю p
for x in range(p):
    y_squared = (pow(x, 3, p) + a * x + b) % p
    if numbertheory.is_square_mod(y_squared, p):
        y = numbertheory.square_root_mod_prime(y_squared, p)
        point = ellipticcurve.Point(curve, x, y)
        break
else:
    raise ValueError("Не удалось найти точку на кривой")

print(f"Используемая кривая: y^2 = x^3 + {a}x + {b} над Z_{p}")
print(f"Точка P: ({point.x()}, {point.y()})")

try:
    order = find_point_order(curve, point, p)
    print(f"Порядок точки P: {order}")
except TimeoutError as e:
    print(e)
```

```python
import time
from sympy import randprime, isprime
from ecpy.curves import Curve, Point

def find_suitable_curve(max_p=1000, max_attempts=100):
    """Подбирает эллиптическую кривую с небольшим порядком точки."""
    for _ in range(max_attempts):
        # Выбираем небольшое простое p
        p = randprime(2, max_p)
        
        # Пробуем случайные коэффициенты a и b
        a = 0
        while True:
            a = (a + 1) % p
            # Уравнение y² = x³ + a*x + b должно быть гладким (без кратных корней)
            # 4a³ + 27b² ≠ 0 mod p
            for b in range(1, p):
                if (4 * a**3 + 27 * b**2) % p != 0:
                    curve = Curve(a, b, p)
                    # Ищем точку на кривой
                    for x in range(p):
                        y_sq = (x**3 + a * x + b) % p
                        # Проверяем, является ли y_sq квадратичным вычетом
                        if pow(y_sq, (p - 1) // 2, p) == 1:
                            # Находим y
                            y = pow(y_sq, (p + 1) // 4, p)  # Только для p ≡ 3 mod 4!
                            point = Point(x, y, curve)
                            return curve, point
    return None, None

def brute_force_order(curve, point):
    """Находит порядок точки полным перебором."""
    if point.is_infinity:
        return 1
    Q = point
    order = 1
    while not Q.is_infinity:
        Q = Q + point
        order += 1
    return order

# Подбираем кривую и точку
curve, point = find_suitable_curve(max_p=100)
if curve and point:
    print(f"Кривая: y² = x³ + {curve.a}x + {curve.b} над F_{curve.field}")
    print(f"Точка: ({point.x}, {point.y})")

    # Измеряем время поиска порядка
    start_time = time.time()
    order = brute_force_order(curve, point)
    elapsed_time = time.time() - start_time

    print(f"Порядок точки: {order}")
    print(f"Время вычисления: {elapsed_time:.2f} сек")
else:
    print("Не удалось подобрать подходящую кривую.")
```
